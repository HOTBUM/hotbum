```python
import matplotlib.pyplot as plt

import seaborn as sns
import numpy as np
```

```python
# 화면 스타일 설정하기

sns.set_style("whitegrid")
```

```python
#인덱스 명 바꾸기
df = df.rename_axis(index="Stock")
# 딕셔너리 명칭 바꾸기
stock_name_mapping = {
    'KOSPI': 'KOSPI',
    '네이버': 'NAVER',
    '카카오': 'KAKAO',
    '카카오게임즈': 'KAKAO Games',
    '카카오뱅크': 'KAKAO Bank',
    '카카오페이': 'KAKAO Pay'
}

# 영문 종목명으로 바꾸기
df = df.set_index(df.index.map(stock_name_mapping))

df
```

```python
# 변수명 영어로 바꾸기
df.columns=['Date','Returns']
```

```python
sns.boxplot(x = 'Stock', y = 'Returns', data = df)

plt.show()
```

```python
sns.boxplot(x = 'Stock', y = 'Returns', data = df, showfliers=False)
plt.xticks(rotation=45, ha='right', wrap=True)
plt.show()
```

```python
flierprops = dict(marker='o', markerfacecolor='Black', markersize=3,
                  linestyle='none')
sns.color_palette(palette='cividis')

sns.boxplot(x = 'Stock', y = 'Returns', data = df, palette='cividis', flierprops=flierprops)
plt.xticks(rotation=20, ha='right', wrap=True)
plt.title('KOREA IT Coporation')
plt.savefig('20250407김학범.png', dpi=500)
plt.show()
```
